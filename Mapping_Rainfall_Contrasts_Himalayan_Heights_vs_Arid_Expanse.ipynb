```python
import matplotlib.pyplot as plt

# Coordinates for Himachal Pradesh (HP) and Rajasthan
hp_coords = (31.1048, 77.2664)  # Shimla, Himachal Pradesh
rajasthan_coords = (27.0238, 74.2179)  # Jaipur, Rajasthan

# Rainfall values (for illustration)
hp_rainfall = 800  # Hypothetical higher rainfall in Himachal Pradesh
rajasthan_rainfall = 200  # Hypothetical lower rainfall in Rajasthan

# Create a basic map
plt.figure(figsize=(8, 6))
plt.scatter(hp_coords[1], hp_coords[0], s=hp_rainfall, c='blue', label='Himachal Pradesh')
plt.scatter(rajasthan_coords[1], rajasthan_coords[0], s=rajasthan_rainfall, c='red', label='Rajasthan')

# Add labels
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Rainfall Comparison: Himachal Pradesh vs Rajasthan')
plt.legend()

# Show the plot
plt.show()
```
