```python
x =[]
JAN =[]
FEB =[]
MAR =[]
APR =[]
for i in range(len(A)):
  print(A[i])
  x.append(A[i][0])
  JAN.append(A[i][1])
  FEB.append(A[i][2])
  MAR.append(A[i][3])
  APR.append(A[i][4])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# # Data for plotting
# x = np.arange(0.0, 2.0, 0.01)
# y = 1 + np.sin(2 * np.pi * t)
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(x, JAN , c='r', label="JANUARY")
ax.plot(x, FEB , c='r', label="FEBRUARY")
ax.plot(x, MAR , c='r', label="MARCH")
ax.plot(x, APR , c='r', label="APRIL")
ax.legend(loc='upper right')


ax.set(xlabel='Relative Area', ylabel='Relative Height',
       title='Snow Cover Hypsometric curve')
ax.grid()

fig.savefig("test.png")
plt.show()
```
