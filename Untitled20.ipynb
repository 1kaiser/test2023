```python
from osgeo import gdal

def extract_pixel_values(raster_path, latitudes, longitudes, band_index=1):
    dataset = gdal.Open(raster_path)
    band = dataset.GetRasterBand(band_index)
    transform = dataset.GetGeoTransform()

    pixel_values = []

    for lat, lon in zip(latitudes, longitudes):
        x_coord = int((lon - transform[0]) / transform[1])
        y_coord = int((lat - transform[3]) / transform[5])

        pixel_value = band.ReadAsArray(x_coord, y_coord, 1, 1)[0, 0]
        pixel_values.append(pixel_value)

    dataset = None
    return pixel_values

# Example usage
latitudes = lat_1_buf
longitudes = lon_1_buf

# Change the band_index argument to 4 for band 4
raster_path = "/content/MOD09A1.061_sur_refl_b04_doy2023001_aid0001.tif"
pixel_values_b4 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)

# Combine the arrays into a single NumPy array
point_cloud_array = np.column_stack((latitudes, longitudes, height_1_buf, pixel_values_b4))

# Save the NumPy array to a file using np.save()
np.save('point_cloud.npy', point_cloud_array)

import numpy as np
import matplotlib.pyplot as plt

# Your data array
data = point_cloud_array

# Extract third and fourth columns
y = data[:, 2]          # height
b4 = data[:, 3]/1000  # reflectance for band 4

# Create a scatter plot for band 4
plt.scatter(b4, y, s=1, c='cyan', alpha=0.3, label='band 4')  # Adjust s, c, and alpha as needed
plt.xlabel('reflectance for band 4')
plt.ylabel('height')
plt.title('Scatter Plot of height vs reflectance for band 4')
plt.xlim(-0.01, 1.6)  # Set x-axis limits
plt.legend()  # Show legend

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genpareto

# Simulated data for rainfall, soil moisture, and temperature variation
np.random.seed(0)
rainfall = np.random.normal(50, 10, 100)  # Simulated rainfall data (e.g., in mm)
soil_moisture = np.random.normal(0.4, 0.1, 100)  # Simulated soil moisture data
temp_variation = np.random.normal(5, 2, 100)  # Simulated temperature variation data (Tmax - Tmin)

# Fit generalized Pareto distributions to the exceedances
rainfall_exceedances = rainfall - rainfall.mean()
soil_moisture_exceedances = soil_moisture - soil_moisture.mean()
temp_variation_exceedances = temp_variation - temp_variation.mean()

# Fitting generalized Pareto distributions
rainfall_params = genpareto.fit(rainfall_exceedances[rainfall_exceedances > 0])
soil_moisture_params = genpareto.fit(soil_moisture_exceedances[soil_moisture_exceedances > 0])
temp_variation_params = genpareto.fit(temp_variation_exceedances[temp_variation_exceedances > 0])

# Generate synthetic data points from the fitted distributions
n_samples = 1000
rainfall_samples = genpareto.rvs(*rainfall_params, size=n_samples)
soil_moisture_samples = genpareto.rvs(*soil_moisture_params, size=n_samples)
temp_variation_samples = genpareto.rvs(*temp_variation_params, size=n_samples)

# Plot the angular densities on a two-dimensional simplex
fig, ax = plt.subplots()
ax.plot(
    rainfall_samples / (rainfall_samples + soil_moisture_samples + temp_variation_samples),
    soil_moisture_samples / (rainfall_samples + soil_moisture_samples + temp_variation_samples),
    'o', markersize=5, alpha=0.5
)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xlabel('Normalized Rainfall')
ax.set_ylabel('Normalized Soil Moisture')

plt.title("Angular Densities")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed
np.random.seed(0)

# Number of samples
n_samples = 100

# Generate synthetic data for base flow and normal rainfall
base_flow = np.random.normal(10, 2, n_samples)
normal_rainfall = np.random.normal(30, 5, n_samples)

# Generate synthetic data for solid temperature, moisture, and rainfall
solid_temperature = np.random.normal(25, 5, n_samples)
moisture = np.random.uniform(0, 1, n_samples)
rainfall = np.random.exponential(scale=10, size=n_samples)

# Define thresholds
threshold_base_flow = 8
threshold_normal_rainfall = 35
threshold_temperature = 30
threshold_moisture = 0.8
threshold_rainfall = 20

# Identify extreme events
extreme_events_base_flow = np.where(base_flow < threshold_base_flow)[0]
extreme_events_normal_rainfall = np.where(normal_rainfall > threshold_normal_rainfall)[0]
joint_extreme_events_mask = np.zeros(n_samples, dtype=bool)
joint_extreme_events_mask[extreme_events_base_flow] = True
joint_extreme_events_mask[extreme_events_normal_rainfall] &= True

extreme_events_temp_moisture_rain = np.where(
    (solid_temperature > threshold_temperature) &
    (moisture > threshold_moisture) &
    (rainfall > threshold_rainfall)
)[0]

# Transform base flow and normal rainfall data into radial and angular components
r = np.sqrt(base_flow**2 + normal_rainfall**2)
theta = np.arctan2(normal_rainfall, base_flow)

# Calculate radial threshold
quantile_threshold = 0.9
threshold_r0 = np.quantile(r, quantile_threshold)
extreme_events_r0 = np.where(r > threshold_r0)[0]

# Plotting
plt.figure(figsize=(18, 12))

# Base Flow vs. Normal Rainfall
plt.subplot(231)
plt.scatter(base_flow, normal_rainfall, color='b', label='Data Points')
plt.scatter(base_flow[extreme_events_base_flow], normal_rainfall[extreme_events_base_flow], color='r', marker='o', label='Extreme Base Flow')
plt.scatter(base_flow[extreme_events_normal_rainfall], normal_rainfall[extreme_events_normal_rainfall], color='g', marker='s', label='Extreme Normal Rainfall')
plt.scatter(base_flow[joint_extreme_events_mask], normal_rainfall[joint_extreme_events_mask], color='m', marker='^', label='Joint Extreme Events')
plt.xlabel('Base Flow')
plt.ylabel('Normal Rainfall')
plt.legend()

# Radial vs. Angular Components
plt.subplot(232)
plt.scatter(r, theta, color='b', label='Data Points')
plt.scatter(r[extreme_events_r0], theta[extreme_events_r0], color='r', marker='o', label=f'Extreme Events (r > {threshold_r0:.2f})')
plt.xlabel('Radial Component (r)')
plt.ylabel('Angular Component (θ)')
plt.legend()

# Solid Temperature
plt.subplot(233)
plt.plot(solid_temperature, label='Solid Temperature')
plt.axhline(y=threshold_temperature, color='r', linestyle='--', label='Threshold')
plt.scatter(extreme_events_temp_moisture_rain, solid_temperature[extreme_events_temp_moisture_rain], color='r', label='Extreme Events')
plt.xlabel('Time')
plt.ylabel('Solid Temperature')
plt.legend()

# Moisture
plt.subplot(234)
plt.plot(moisture, label='Moisture')
plt.axhline(y=threshold_moisture, color='r', linestyle='--', label='Threshold')
plt.scatter(extreme_events_temp_moisture_rain, moisture[extreme_events_temp_moisture_rain], color='r', label='Extreme Events')
plt.xlabel('Time')
plt.ylabel('Moisture')
plt.legend()

# Rainfall
plt.subplot(235)
plt.plot(rainfall, label='Rainfall')
plt.axhline(y=threshold_rainfall, color='r', linestyle='--', label='Threshold')
plt.scatter(extreme_events_temp_moisture_rain, rainfall[extreme_events_temp_moisture_rain], color='r', label='Extreme Events')
plt.xlabel('Time')
plt.ylabel('Rainfall')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet

# Generate a grid of points on a triangle (simplex).
def generate_triangle_grid(n):
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)
    X, Y = np.meshgrid(x, y)
    mask = X + Y <= 1
    X, Y = X[mask], Y[mask]
    return X, Y

# Define the Dirichlet density function.
def dirichlet_density(a, w1, w2):
    """
    Returns the Dirichlet density for given parameters a and points w1 and w2.
    w3 is inferred as w3 = 1 - w1 - w2.
    """
    # Infer w3 from w1 and w2
    w3 = 1 - w1 - w2

    # Create the input array with three columns: w1, w2, and w3
    w = np.vstack([w1.ravel(), w2.ravel(), w3.ravel()]).T

    # Compute Dirichlet density
    pdf_values = dirichlet.pdf(w, a)

    # Reshape the PDF values to match w1 and w2's shape
    return pdf_values.reshape(w1.shape)

# Parameters for the Dirichlet distribution.
dirichlet_params = [(4, 4, 4), (0.25, 0.25, 0.25), (0.25, 0.75, 1.2)]

# Generate the triangular grid.
W1, W2 = generate_triangle_grid(100)
W3 = 1 - W1 - W2

# Plotting.
fig, axs = plt.subplots(1, 3, figsize=(15, 5), subplot_kw={'projection': '3d'})

fig.suptitle("Dirichlet Model Densities")
for ax, params in zip(axs, dirichlet_params):
    Z = dirichlet_density(params, W1, W2)
    ax.scatter(W1, W2, W3, c=Z, cmap="viridis")
    ax.set_title(f"α = {params}")

plt.show()
```
